```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://www.imdb.com/chart/top?ref_=nv_mv_250")
```

```python
soup = BeautifulSoup(result.content, 'lxml')
```

```python
trs = soup.findAll('table')[0].findAll('tr')
```
